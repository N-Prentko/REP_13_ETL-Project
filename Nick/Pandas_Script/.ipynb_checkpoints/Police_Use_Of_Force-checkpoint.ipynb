```python
import pandas as pd
import datetime as dt
import numpy as np
```

```python
#load csv into dataframe
load_csv = ("../Source_Data/Police_Use_of_Force_Source_Data.csv")
```

```python
source_data_df = pd.read_csv(load_csv)
source_data_df.head()
```

```python
edited_source_data_df = source_data_df.drop(columns = {"X", "Y", "Is911Call", "ForceReportNumber",
                                  "SubjectRoleNumber", "TotalCityCallsForYear",
                                  "TotalPrecinctCallsForYear", "TotalNeighborhoodCallsForYear",
                                  "CenterGBSID", "CenterLatitude", "CenterLongitude",
                                  "CenterX", "CenterY", "DateAdded", "OBJECTID"})

edited_source_data_df.head(5)
```

```python
renamed_columns_df = edited_source_data_df.rename(columns={'PoliceUseOfForceID': 'police_use_of_force_id',
                                                          'CaseNumber': 'case_number', 'ResponseDate': 'response_date',
                                                          'PrimaryOffense': 'primary_offense', 'SubjectInjury': 'subject_injury',
                                                          'SubjectRole': 'subject_role', 'ForceType': 'police_use_of_force_type',
                                                          'ForceTypeAction': 'force_type_action', 'Race': 'subject_race', 
                                                          'Sex': 'subject_sex', 'EventAge': 'subject_age', 'TypeOfResistance': 'type_of_resistance',
                                                          'Problem': 'problem', 'Precinct': 'precinct', 'Neighborhood': 'neighborhood'})

renamed_columns_df.head(5)
```

```python
renamed_columns_df["nbhd_key"] = ""
```

```python
renamed_columns_df.head(5)
```

```python
edited_renamed_columns_df = renamed_columns_df.rename(columns={'Precinct': 'precinct',
                                                               'Problem': 'problem',
                                                              'Neighborhood': 'neighborhood'})
```

```python
edited_renamed_columns_df.head(5)
```

```python
# Bring in Swati's MLPS_Neighborhoods.csv
load_second_csv = ("../target_files/MLPS_Neighborhoods.csv")
nbhd_keys_df = pd.read_csv(load_second_csv)
nbhd_keys_df.head(5)
```

```python
nbhd_keys_df_edited = nbhd_keys_df.rename(columns={'NBHD_KEY': 'nbhd_key',
                                                  'name': 'neighborhood'})
```

```python
nbhd_keys_df_edited.head(5)
```

```python
left_join_df = pd.merge(edited_renamed_columns_df, nbhd_keys_df_edited, on = 'neighborhood', how= 'left')
left_join_df
```

```python
left_join_df.drop(columns= {"nbhd_key"})
```

```python
rearranged_df = left_join_df[['police_use_of_force_id','response_date',
                              'case_number','problem','subject_race',
                             'subject_sex', 'subject_age','subject_role', 'primary_offense',
                             'type_of_resistance', 'police_use_of_force_type',
                             'force_type_action', 'subject_injury', 'neighborhood_id',
                             'neighborhood', 'precinct']]
rearranged_df
```

```python
rearranged_df.dtypes
```

```python
rearranged_df.response_date=pd.to_datetime(rearranged_df.response_date)
rearranged_df.head(5)
```

```python
near_final_police_use_of_force_df = rearranged_df
```

```python
# Check to see if subject_age field is int
near_final_police_use_of_force_df.dtypes
```

```python
# Change all NaN(s) in the subject_age field to 0
near_final_police_use_of_force_df['subject_age'] = near_final_police_use_of_force_df['subject_age'].replace(np.nan, 0)
```

```python
# Confirm task from cell above was completed
near_final_police_use_of_force_df
```

```python
# Convert subject_age field to int64
near_final_police_use_of_force_df['subject_age'] = near_final_police_use_of_force_df['subject_age'].astype(np.int64)
```

```python
# Examine subject_age field to confirm the absence of decimals
near_final_police_use_of_force_df
```

```python
# Double check that the subject_age field is int64
near_final_police_use_of_force_df.dtypes
```

```python
# Adjust punctuation of neighborhood name to match MLS_Neighborhoods.csv
PUOF_final_df = near_final_police_use_of_force_df.replace(to_replace = "St. Anthony West", value ="ST ANTHONY WEST")
PUOF_final_df = near_final_police_use_of_force_df.replace(to_replace = "St. Anthony East", value ="ST ANTHONY EAST")
PUOF_final_df = near_final_police_use_of_force_df.replace(to_replace = "Humboldt Industrial Area", value ="HUMBOLDT INDUST AREA")
PUOF_final_df = near_final_police_use_of_force_df.replace(to_replace = "Camden Industrial", value ="CAMDEN INDUSTRIAL AREA")        
PUOF_final_df = near_final_police_use_of_force_df.replace(to_replace = "CARAG", value ="South Uptown")
PUOF_final_df = near_final_police_use_of_force_df.replace(to_replace = "Bryn - Mawr", value ="Bryn Mawr")
PUOF_final_df = near_final_police_use_of_force_df.replace(to_replace = "Columbia Park", value ="Columbia")
PUOF_final_df = near_final_police_use_of_force_df.replace(to_replace = "Mid - City Industrial", value ="Mid City Industrial Ar")
PUOF_final_df = near_final_police_use_of_force_df.replace(to_replace = "Nicollet Island - East Bank", value ="Nicollet Is/East Bank")
PUOF_final_df = near_final_police_use_of_force_df.replace(to_replace = "University of Minnesota", value ="University")
PUOF_final_df = near_final_police_use_of_force_df.replace(to_replace = "McKinley", value ="Mckinley")
PUOF_final_df['neighborhood']= PUOF_final_df['neighborhood'].str.upper()
```

```python
# Examine edited df
PUOF_final_df
```

```python
# Create csv out of df
PUOF_final_df.to_csv("../target_files/cleaned_data_police_use_of_force.csv", index=False, header=True)
```
